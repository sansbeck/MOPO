```python
import numpy as np
import matplotlib.pyplot as plt

# Define the objective function
def f(x, y):
    return 2 * x + 3 * y

# Define the constraints
def constraint1(x):
    return 6 - 2 * x

def constraint2(x):
    return (8 - x) / 2

# Create a range of x values
x = np.linspace(0, 5, 400)

# Calculate corresponding y values for the constraints
y1 = constraint1(x)
y2 = constraint2(x)

# Plot the constraints
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='2x + y ≤ 6', linestyle='--', color='red')
plt.plot(x, y2, label='x + 2y ≤ 8', linestyle='--', color='green')

# Plot the feasible region (shaded)
plt.fill_between(x, 0, np.minimum(y1, y2), where=(x >= 0) & (x <= 4), color='gray', alpha=0.5, label='Feasible Region')

# Plot the optimal solution
plt.plot(2, 2, 'ro', label='Optimal Solution (2, 2)')

# Label the axes and add a legend
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 5)
plt.ylim(0, 4)
plt.legend()

# Show the plot
plt.title('Optimal Solution and Constraints')
plt.grid(True)
plt.show()
```
